```python
#importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#carregando os dados
df = pd.read_csv ('/content/dados(1).csv')
```

```python
df
```

```python
#Calculando valores mínimos, máximos e a média da pontuação de cada matéria
df.describe().loc[['min', 'max', 'mean']]
# aqui foi feito o uso da função describe() para descrever os dados, e também o uso da função loc(), pois queria apenas algumas linha
# específicas.
```

```python
#Agrupando o banco de dados a partir da raça/etnia dos alunos;
df2 = df.groupby('race/ethnicity').count()
df2
```

```python
# Plotando Gráfico do df2
df2['math score'].plot(kind='barh',color ='pink', title = 'Média em Math de cada grupo')
#primeiro foi especificado o retorno de valores que era desejado, e depois com o código plot gerando o grafico
# e o kind para especificar o tipo de gráfico
```

```python
# Agrupando o banco de dados a partir do nível de escolaridade dos pais dos alunos
df3 = df.groupby('parental level of education').count()
df3
```

```python
# PLotando o gráfico do df3
df3['writing score'].plot(kind='barh',color ='pink', title = 'Média em writing de cada grupo')
```

```python
#Agrupando o banco de dados a partir da participação em um curso preparatório para os testes
df4 = df.groupby('test preparation course').count()
df4
```

```python
# PLotando o gráfico do df4
df4['reading score'].plot(kind='barh',color ='pink', title = 'Média em reading de cada grupo')
```
