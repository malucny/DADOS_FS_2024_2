```python
#importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#carregando os dados
df = pd.read_csv ('/content/dados(1).csv')
```

```python
df
```

```python
#Calculando valores mÃ­nimos, mÃ¡ximos e a mÃ©dia da pontuaÃ§Ã£o de cada matÃ©ria
df.describe().loc[['min', 'max', 'mean']]
# aqui foi feito o uso da funÃ§Ã£o describe() para descrever os dados, e tambÃ©m o uso da funÃ§Ã£o loc(), pois queria apenas algumas linha
# especÃ­ficas.
```

```python
#Agrupando o banco de dados a partir da raÃ§a/etnia dos alunos;
df2 = df.groupby('race/ethnicity').count()
df2
```

```python
# Plotando GrÃ¡fico do df2
df2['math score'].plot(kind='barh',color ='pink', title = 'MÃ©dia em Math de cada grupo')
#primeiro foi especificado o retorno de valores que era desejado, e depois com o cÃ³digo plot gerando o grafico
# e o kind para especificar o tipo de grÃ¡fico
```

```python
# Agrupando o banco de dados a partir do nÃ­vel de escolaridade dos pais dos alunos
df3 = df.groupby('parental level of education').count()
df3
```

```python
# PLotando o grÃ¡fico do df3
df3['writing score'].plot(kind='barh',color ='pink', title = 'MÃ©dia em writing de cada grupo')
```

```python
#Agrupando o banco de dados a partir da participaÃ§Ã£o em um curso preparatÃ³rio para os testes
df4 = df.groupby('test preparation course').count()
df4
```

```python
# PLotando o grÃ¡fico do df4
df4['reading score'].plot(kind='barh',color ='pink', title = 'MÃ©dia em reading de cada grupo')
```
