```python
# Plotando um gráfico de barras (bar) que exiba a quantidade de
# funcionários em cada setor (coluna Department)
df2 = df.groupby('Department')[['EmployeeNumber']].count()
df2
```

```python
df2.plot(kind = 'bar')
```

```python
#Plotando um gráfico de linha (line) que exiba a quantidade de
#funcionários para cada nível de escolaridade (coluna Education);
df3 = df.groupby('Education')[['EmployeeNumber']].count()
df3
```

```python
df3.plot(kind = 'line')
```
